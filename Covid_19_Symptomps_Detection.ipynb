#Data Preprocessing

```python
df_OH = pd.get_dummies(df, drop_first=True) # to avoid dummy trap
df_OH.head()
```

```python
corr = df_OH.corr()
corr.style.background_gradient(cmap='coolwarm',axis=None)
#next feature
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(18, 10))

sns.heatmap(corr);
```

```python
df_conv = np.where(df.values == 'Yes',1,0)
df_conv = pd.DataFrame(df_conv, columns=df.columns)
df_conv
```

```python
X = df_conv.iloc[:,:-1]
y = df_conv.iloc[:,-1]
```

#Handle imbalanced class using Random Undersampling

```python
from collections import Counter
from imblearn.under_sampling import RandomUnderSampler 

print('Original dataset shape %s' % Counter(y))

tl = RandomUnderSampler(random_state=10)
X_res, y_res = tl.fit_resample(X, y)
print('Resampled dataset shape %s' % Counter(y_res))
# Resampled dataset shape Counter({1: 897, 0: 100})
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, train_size = 0.7, random_state=0)
```

#Building Model

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score ,precision_score,recall_score,f1_score

decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train,y_train) 
Y_pred = decision_tree.predict(X_test) 

accuracy_dt=round(accuracy_score(y_test,Y_pred)* 100, 2)
acc_decision_tree = round(decision_tree.score(X_train, y_train) * 100, 2)
cm = confusion_matrix(y_test, Y_pred)
accuracy = accuracy_score(y_test,Y_pred)
precision = precision_score(y_test, Y_pred,average='micro')
recall = recall_score(y_test, Y_pred,average='micro')
f1 = f1_score(y_test,Y_pred,average='micro')

print('Confusion matrix for DecisionTree\n',cm)
print('accuracy_DecisionTree: %.3f' %accuracy)
print('precision_DecisionTree: %.3f' %precision)
print('recall_DecisionTree: %.3f' %recall)
print('f1-score_DecisionTree : %.3f' %f1)
```

```python
from sklearn import metrics
cm = confusion_matrix(y_test, Y_pred)
p = sns.heatmap(pd.DataFrame(cm), annot=True, cmap="YlGnBu" ,fmt='g')
plt.title('Confusion matrix for Decision Tree')
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize = (15,15))
plot_tree(decision_tree.fit(X_train, y_train) ,filled=True)
plt.show()
```

```python
from sklearn.svm import LinearSVC

linear_svc = LinearSVC(max_iter=4000)
linear_svc.fit(X_train, y_train)
Y_pred = linear_svc.predict(X_test)
accuracy_svc=round(accuracy_score(y_test,Y_pred)* 100, 2)
acc_linear_svc = round(linear_svc.score(X_train, y_train) * 100, 2)
cm = confusion_matrix(y_test, Y_pred)
accuracy_SVM = accuracy_score(y_test,Y_pred)
precision = precision_score(y_test, Y_pred,average='micro')
recall = recall_score(y_test, Y_pred,average='micro')
f1 = f1_score(y_test,Y_pred,average='micro')

print('Confusion matrix for SVC\n',cm)
print('accuracy_SVC: %.3f' %accuracy)
print('precision_SVC: %.3f' %precision)
print('recall_SVC: %.3f' %recall)
print('f1-score_SVC : %.3f' %f1)
```
